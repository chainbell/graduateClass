```python
#군집분석을 위한 라이브러리
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
# 군집 수를 설정합니다.
cluster_nums = [3, 4, 5, 6, 7]

for num_clusters in cluster_nums:
    # K-Means 모델을 생성하고 학습
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    kmeans.fit(fish_input)

    # 클러스터 라벨
    cluster_labels = kmeans.labels_

    # 군집 분석 결과와 실제 라벨을 비교하여 Adjusted Rand Index를 계산
    ari = adjusted_rand_score(fish_target, cluster_labels)

    # 결과 출력
    print(f"Number of Clusters: {num_clusters}, Adjusted Rand Index: {ari}")
```
